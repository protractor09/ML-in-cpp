```python
%%manim -ql LogisticRegressionPlot

import numpy as np
from manim import *

class LogisticRegressionPlot(Scene):
    def construct(self):
        time_rained=np.array([1,2,4,0,3,5,1.5,4.5,0.5,6])
        flood_lvl=np.array([0,5,12,0,10,20,2,18,0,25])

        y_data = np.array([1 if f > 1 else 0 for f in flood_lvl])


        w = 2
        b = -5
        axes = Axes(
            x_range=[-1, 7],
            y_range=[-0.1, 1.1],
            x_length=8,
            y_length=4,
            axis_config={"include_tip": False}
        )
        labels = axes.get_axis_labels(x_label="Time Rained", y_label="Flood Level")
        def sig(z):
          return 1 / (1 + np.exp(-(w * z + b)))

        dots = VGroup()
        pos_dot=VGroup()
        for x_val, y_val in zip(time_rained, y_data):
            dot = Dot(point=axes.c2p(x_val, y_val), color=RED)
            dots.add(dot)

        dot1 = Dot(point=axes.c2p(0, 0.7), color=YELLOW)
        dot2= Dot(point=axes.c2p(2.9,0),color=YELLOW )
        dotn=Dot(point=axes.c2p(2.9,0.7),color=WHITE)
        pos_dot.add(dot1)
        pos_dot.add(dot2)
        pos_dot.add(dotn)

        vline1 = Line(start=dot1, end=dotn, color=YELLOW)
        vline2 = Line(start=dotn, end=dot2, color=YELLOW)

        txt=Text("Flood Level = 7cm",font_size=30)
        txt.move_to(DOWN*2.5)
        txt2=Text("Time Rained = 2.9hrs",font_size=30)
        txt2.move_to(DOWN*3.125)


        curve=axes.plot(sig)


        self.play(Create(axes),Write(labels))
        self.play(Create(curve),run_time=2)
        self.play(FadeIn(dots))
        self.play(FadeIn(pos_dot))
        self.play(Create(vline1))
        self.play(Create(vline2))
        self.play(Create(txt))
        self.play(Create(txt2))
        self.wait(2)
```
